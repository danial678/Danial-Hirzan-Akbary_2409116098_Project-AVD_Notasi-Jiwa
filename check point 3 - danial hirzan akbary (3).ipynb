### Outliers

Melakukan pengecekan pada Outliers

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

melakukan imputasi

```python
columns_to_impute = ["Age", "Hours per day","BPM"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

melakukan pengecekan ulang

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## Import Libary

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Pie Chart

### Menggunakan Pie Chart untuk mengetahui platfrom apa yang sering digunakan

```python
platform_counts = df['Primary streaming service'].value_counts()
plt.figure(figsize=(6, 8))
plt.pie(platform_counts, labels=platform_counts.index, autopct='%1.1f%%', colors=sns.color_palette("viridis", len(platform_counts)), startangle=140)

plt.title("Proporsi Pengguna Tiap Platform Streaming Musik", fontsize=14)
plt.show()
```

## Barchart

### Menggunakan barchart untuk mendistribusikan rata rata tingkat stress berdasarkan genre

```python
stress_avg = df.groupby('Fav genre')[['Anxiety', 'Depression', 'Insomnia', 'OCD']].mean()

plt.figure(figsize=(12,6))
stress_avg.plot(kind='bar', colormap='viridis', alpha=0.9, width=0.85)
plt.title("Rata-rata Tingkat Stres Berdasarkan Genre Musik Favorit", fontsize=14)
plt.xlabel("Genre Musik", fontsize=12)
plt.ylabel("Tingkat Stres (Rata-rata)", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Kategori Stres")

plt.show()
```

### Menggunakan barchart untuk melihat disribusi efek musik dalam kesehatan mental

```python
music_effects_avg = df.groupby("Fav genre")["Music effects"].value_counts().unstack()

# Plot bar chart
plt.figure(figsize=(12, 6))
music_effects_avg.plot(kind="bar", colormap="viridis", alpha=0.9, width=1)

# Tambahkan judul dan label
plt.title("Distribusi Efek Musik Berdasarkan Genre Musik Favorit", fontsize=14)
plt.xlabel("Genre Musik", fontsize=12)
plt.ylabel("Jumlah Responden", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Music Effects")

# Tampilkan plot
plt.show()
```

## Heatmap

### Menggunakan heatmap untuk melihat korelasi skor gangguan mentap berdasarkan genre

```python
pivot_table = df.pivot_table(index="Fav genre", values=["Anxiety", "Depression", "OCD", "Insomnia"], aggfunc="mean")

# Membuat heatmap
plt.figure(figsize=(8,5))
sns.heatmap(pivot_table, cmap="coolwarm", annot=True, fmt=".2f")

plt.title("Rata-rata Skor Gangguan Mental Berdasarkan Genre Musik Favorit")
plt.xlabel("Gangguan Mental")
plt.ylabel("Genre Musik")
plt.show()
```
